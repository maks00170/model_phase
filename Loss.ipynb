```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('metrics1.csv')
print('min train_loss = ', df.groupby('epoch').sum().train_loss_epoch.min())
print('min valid_loss = ', df.groupby('epoch').sum().valid_loss.min())

plt.title('1 вариант')
plt.plot(df.groupby('epoch').sum().valid_loss, label = 'valid_loss')
plt.plot(df.groupby('epoch').sum().train_loss_epoch, label = 'train_loss')
plt.xlabel('epochs')
plt.ylabel('L1loss')
plt.legend();
```

```python
df = pd.read_csv('metrics2.csv')

print('min train_loss = ', df.groupby('epoch').sum().train_loss_epoch[:15].min())
print('min valid_loss = ', df.groupby('epoch').sum().valid_loss[:15].min())

plt.title('Вариант 2')
plt.plot(df.groupby('epoch').sum().valid_loss[:15], label = 'valid_loss')
plt.plot(df.groupby('epoch').sum().train_loss_epoch[:15], label = 'train_loss')
plt.xlabel('epochs')
plt.ylabel('L1loss')
plt.legend();
```
